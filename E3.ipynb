```python
import cartopy.crs as ccrs
import cartopy
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import numpy as np
# these modules will be used later
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
```

## Excercise 1: Mapping Antarctica and Southern Ocean 

```python
plt.figure()
ax = plt.axes(projection=ccrs.Stereographic(central_latitude=-90,
                                            central_longitude=0))
ax.set_extent([-180, 180, -90, -60], crs=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cfeature.LAND, facecolor='lightgray')

gl = ax.gridlines(draw_labels=True, linestyle='--', linewidth=0.5, alpha=0.7)

gl.xlabel_style = {'rotation': 0}
```

```python
## Figure 1: Map of Antarctica and the surrounding Southern Ocean.
```

## Excercise 2: Mapping the South Atlantic Ocean and annotating four coastal cities

```python
# a map of the South Atlantic Ocean, with  2 South American and  2 African cities 
geolocator = Nominatim(user_agent='educational') # nominatim is a free service and requires an identificative agent
extent = [-70, 30, -50, -20]
central_lon = np.mean(extent[:2])
central_lat = np.mean(extent[2:])

place = ['Walvis Bay','Cape Town','Rio de Janeiro','Montevideo']
address = []
for p in place:
    loc = geolocator.geocode(p,language="en")
    address.append(loc)
print(address)

plt.figure(figsize=(12, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_extent(extent, crs=ccrs.PlateCarree())

ax.text(-20, -40, 'South Atlantic Ocean',
        transform=ccrs.PlateCarree(),
        fontsize=14,
        fontstyle='italic',
        ha='center')

ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
ax.add_feature(cfeature.COASTLINE, linewidth=0.8)

gl = ax.gridlines(draw_labels=True, linestyle='--', linewidth=0.5, alpha=0.7)
gl.top_labels = False
gl.right_labels = False
gl.xlabel_style = {'rotation': 0}
for p in range(len(place)):
    ax.plot(address[p].longitude,address[p].latitude,'ko',markersize=5,transform=ccrs.PlateCarree())
    ax.text(address[p].longitude+0.3,address[p].latitude,place[p],transform=ccrs.Geodetic())
```
